```python
import shutil
#Utility function to move corresponding images and labels to the respective folders
def move_files_to_folder(list_of_files, destination_folder):
    for value in list_of_files:
        try:
            shutil.move(value, destination_folder)
        except:
            print(value)
            assert False
```

```python
# importing the module
import json
 
# Opening JSON file
with open('/content/datasets/Kvasir-SEG/kvasir_bboxes.json') as json_file:
    data = json.load(json_file)
 
    # Print the type of data variable
    print("Type:", type(data))
 
    # Print the data of dictionary
    #print("\ncju0qkwl35piu0993l0dewei2:", data['cju0qkwl35piu0993l0dewei2']['bbox'])
for rec in data:
    print(rec)
    for dim, dimvalue in data[rec].items():
      print (dim, dimvalue)
      if dim=='height':
         height = dimvalue
      elif dim=='width':
         width = dimvalue
      elif dim=='bbox':
        print(len(dimvalue))
        length = len(dimvalue)
        i = 0
  
        # Iterating using while loop
        while i < length:
           label = dimvalue[i]['label']
           xmin = dimvalue[i]['xmin']
           ymin = dimvalue[i]['ymin']
           xmax = dimvalue[i]['xmax']
           ymax = dimvalue[i]['ymax']
           
           if i==0: 
             with open('/content/datasets/Kvasir-SEG/labels/'+rec+'.txt', 'w') as file:
                file.write('0'+ ' ' + str(round(((xmin+xmax)/2)/width,4)) + ' ' + str(round(((ymin+ymax)/2)/height,4)) + ' ' + str(round((xmax-xmin)/width,4)) + ' ' + str(round((ymax-ymin)/height,4)) + '\n')
           else:
             with open('/content/datasets/Kvasir-SEG/labels/'+rec+'.txt', 'a') as file:
                file.write('0'+ ' ' + str(round(((xmin+xmax)/2)/width,4)) + ' ' + str(round(((ymin+ymax)/2)/height,4)) + ' ' + str(round((xmax-xmin)/width,4)) + ' ' + str(round((ymax-ymin)/height,4)) + '\n')
           #0,norm x center -> (xmin+xmax/2)/width, norm y center -> (ymin+ymax/2)/height , norm width -> (xmax-xmin) /width, norm height -> (ymax-ymin) /height
           i += 1
       
```
