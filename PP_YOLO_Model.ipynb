```python
import xml.etree.ElementTree as gfg
```

```python
def GenerateXML(filename,height,width,label,xmin,ymin,xmax,ymax,i) :
  root=gfg.Element("annotation")
  
  m1=gfg.Element("folder")
  m1.text="images"
  root.append(m1)

  m2=gfg.Element("filename")
  m2.text=filename+'.jpg' 
  root.append(m2)
  
  m3=gfg.Element("size")
  root.append(m3)

  a1=gfg.SubElement(m3, "width")
  a1.text = width
  a2=gfg.SubElement(m3, "height")
  a2.text = height
  a3=gfg.SubElement(m3, "depth")
  a3.text = '3'
  
  m4 = gfg.Element("object")
  root.append(m4)
  b1=gfg.SubElement(m4, "name")
  b1.text = 'polyp'
  b2=gfg.SubElement(m4, "bndbox")
  
  d1=gfg.SubElement(b2, "xmin")
  d1.text = xmin
  d2=gfg.SubElement(b2, "ymin")
  d2.text = ymin
  d3=gfg.SubElement(b2, "xmax")
  d3.text = xmax
  d4=gfg.SubElement(b2, "ymax")
  d4.text = ymax
  
  
  tree = gfg.ElementTree(root)
  fileName='/content/PaddleDetection/test2.xml'
  with open (fileName, "wb") as files :
    tree.write(files)

# Driver Code
#if __name__ == "__main__":
#	GenerateXML("Catalog.xml")
```

```python
# importing the module
import json
import xml.etree.ElementTree as gfg

# Opening JSON file
with open('/content/Kvasir-SEG-format/kavsir_bboxes.json') as json_file:
    data = json.load(json_file)
 
    # Print the type of data variable
    print("Type:", type(data))
 
    # Print the data of dictionary
    #print("\ncju0qkwl35piu0993l0dewei2:", data['cju0qkwl35piu0993l0dewei2']['bbox'])
for rec in data:
    print(rec)
    for dim, dimvalue in data[rec].items():
      print (dim, dimvalue)
      if dim=='height':
         height = dimvalue
      elif dim=='width':
         width = dimvalue
      elif dim=='bbox':
        print(len(dimvalue))
        length = len(dimvalue)
        i = 0
  
        # Iterating using while loop
        while i < length:
           label = dimvalue[i]['label']
           xmin = dimvalue[i]['xmin']
           ymin = dimvalue[i]['ymin']
           xmax = dimvalue[i]['xmax']
           ymax = dimvalue[i]['ymax']
           
           fileName = rec
               

           if i==0: 
              root=gfg.Element("annotation")
              
              m1=gfg.Element("folder")
              m1.text="images"
              root.append(m1)

              m2=gfg.Element("filename")
              m2.text=fileName+'.jpg' 
              root.append(m2)
              
              m3=gfg.Element("size")
              root.append(m3)

              a1=gfg.SubElement(m3, "width")
              a1.text = str(width)
              a2=gfg.SubElement(m3, "height")
              a2.text = str(height)
              a3=gfg.SubElement(m3, "depth")
              a3.text = '3'

              m4 = gfg.Element("object")
              root.append(m4)
              b1=gfg.SubElement(m4, "name")
              b1.text = 'polyp'
              b2=gfg.SubElement(m4, "bndbox")
              
              d1=gfg.SubElement(b2, "xmin")
              d1.text = str(xmin)
              d2=gfg.SubElement(b2, "ymin")
              d2.text = str(ymin)
              d3=gfg.SubElement(b2, "xmax")
              d3.text = str(xmax)
              d4=gfg.SubElement(b2, "ymax")
              d4.text = str(ymax)

           else:
              m4 = gfg.Element("object")
              root.append(m4)
              b1=gfg.SubElement(m4, "name")
              b1.text = 'polyp'
              b2=gfg.SubElement(m4, "bndbox")
              
              d1=gfg.SubElement(b2, "xmin")
              d1.text = str(xmin)
              d2=gfg.SubElement(b2, "ymin")
              d2.text = str(ymin)
              d3=gfg.SubElement(b2, "xmax")
              d3.text = str(xmax)
              d4=gfg.SubElement(b2, "ymax")
              d4.text = str(ymax)

             
           i += 1
       
    tree = gfg.ElementTree(root)
    fileName='/content/PaddleDetection/Kvasir-SEG-format/Annotations/'+fileName+'.xml'
    with open (fileName, "wb") as files :
      tree.write(files)
```
